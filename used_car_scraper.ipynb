```python
import urllib
import urllib.request
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
```

```python
weburl = 'https://www.usedcarsni.com/search_results.php?make=0&model=0&search_type=1&age_from=0&age_to=0&body_style=0&distance_enabled=0&distance_miles=50&distance_postcode=&doors=999999&fuel_type=0&homepage_search_attr=1&keywords=&mileage_to=0&pagepc0=2&price_from=0&price_to=0&tab_id=0&tax_cost=999999&trans_type=0&user_type=0'
```

```python
def get_the_urls(weburl):
    for i in range(1, 2):
        url_string = weburl[:243] + str(i) + weburl[244:]
        url_page = urllib.request.urlopen(url_string)
    
        soup = BeautifulSoup(url_page, 'lxml')
        
        url_list = []

        for car in soup.find_all('div', {'class':'car-description'}):
            for a in car.find_all('a', href = True):
                found_urls = re.split(r'(https?:\/\/.*?)', a['href'])
                url_list.append(found_urls)
        return url_list
```

```python
url_list = get_the_urls(weburl)
```

```python
urls = []

for i in url_list:
    urls.append(('https://www.usedcarsni.com' + str(i)).replace('[', '').replace(']', '').replace('\'', ''))
```

```python
print(len(urls))
```

```python
def scrape_that_technical_stuff(attribute):
    attribute = []
    
    for link in urls:
        url_page = urllib.request.urlopen(link)
        
        soup = BeautifulSoup(url_page, 'lxml')


        for car in soup.find_all('div', {'class':'technical-info'}):
            pn = car.get_text()
            attribute.append(pn)

    return attribute
```

```python
technical_info = scrape_that_technical_stuff('technical_info')
```

```python
print(technical_info)
```

```python
tech_info = []

for i in technical_info:
    tech_info.append(i.replace('\n', '').replace(
        "$(document).ready(function () {        setPopover('.price-stats-link')    })", '').replace(
        "$(document).ready(function () {                setPopover('.hpi-link')            })", '').replace(
        "$(document).ready(function () {        $('.popover-link').popover({            trigger: isTouchscreenDevice() ? \"click\" : \"hover\"        });    }); ", '').strip())
```
